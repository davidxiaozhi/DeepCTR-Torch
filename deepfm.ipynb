FM 的特征相乘是使用的简化形式

$ab= \dfrac{1}{2} \times [(a + b)^2 - (a^2+b^2)]$


```python
import torch
class FM(torch.nn.Module):
    """input shape: [bs, field_size, embedding_size]
       output shape: [bs, embedding_size]
    """
    def __init__(self):
        super(FM, self).__init__()
        
    def forward(self, x):
        #dim =1 由于输入是三维,求和后所有的field_size 消失, embedding_size 依据 field_size 求和
        square_sum = torch.power(torch.sum(x,dim=1), 2)
        sum_square = torch.sum(x*x,dim=1)
        cross_item = 0.5*(square_sum - sum_square)
        return cross_item
        
```
